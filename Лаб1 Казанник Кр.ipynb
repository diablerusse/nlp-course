# Токенизация в NLP

**Токенизация** – процесс разбиения текста на минимальные единицы (токены): слова, подслова или символы


## 1. Простые методы токенизации

### 1.1. Разделение по пробелам


```python
text = "Hello, world! This is a test."
text.split()
```


### 1.2. Регулярные выражения


```python
import re

text = "Hello, world! This is a test."
tokens = re.findall(r"\w+", text)
tokens
```


**Ограничения простых методов**:  
- Не учитывают пунктуацию
- Не работают с разными языками и морфологией
- Не подходят для серьёзных задач

## 5. Пример предобработки текста

Возьмём новостной текст и применим полный пайплайн предобработки: очистка → токенизация → лемматизация → удаление стоп-слов


```python
import spacy
from nltk.corpus import stopwords
import nltk
import re

nltk.download('stopwords')

# Загрузка модели spaCy для английского
nlp = spacy.load("en_core_web_sm")
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # 1. Приведение к нижнему регистру
    text = text.lower()
    # 2. Удаление пунктуации и цифр
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # 3. Токенизация и лемматизация через spaCy
    doc = nlp(text)
    # 4. Удаление стоп-слов и коротких токенов
    tokens = [token.lemma_ for token in doc if token.text not in stop_words and len(token.text) > 2]
    return tokens

# Пример
sample_text = "The quick brown fox jumps over the lazy dog. This is a simple sentence for preprocessing!"
processed = preprocess_text(sample_text)
print("Оригинал:", sample_text)
print("После предобработки:", processed)
```

Ваш ответ здесь:

1.
Out-of-Vocabulary (OOV) — это проблема, когда языковая модель встречает в тексте слово, которого не было в её словаре во время обучения. Обычно алгоритм не знает, что с ним делать, и просто заменяет его на заглушку (типа токена <UNK>). Из-за этого при переводе или генерации мы теряем некоторые смысловые части, в частности, если это были чьи-то фамилии, редкие термины, заимствования или просто слова с опечатками.

2.
Авторы придумали неординарный и более комплексный подход — они совместили работу со словами и с отдельными буквами.
В их архитектуре основная модель переводит текст на уровне целых слов (это быстрее и лучше улавливает контекст). Но как только система натыкается на то самое неизвестное слово, она передает задачу вспомогательной посимвольной нейросети. Эта сеть генерирует перевод сложного слова буквально по буквам, после чего готовое слово возвращается в основное предложение. В итоге модель может переводить текст без жестких ограничений словаря.

3.
Простые методы: банальное разделение текста по пробелам или с помощью регулярных выражений (работает быстро, но часто ломается на пунктуации).

Пословные и на уровне предложений: когда текст бьется на логические слова и фразы (мы делали это через NLTK, spaCy и Stanza).

Подсловные (subwords): когда слова бьются на частые слоги/сочетания (вроде BPE). Как я понял(а) из лекции, это сейчас главный стандарт для крутых LLM-моделей (тот же токенизатор от HuggingFace).

Посимвольные: текст разбивается на отдельные буквы и знаки.

Послоговые: для языков, где слоги играют важную роль (например, японский).

Гибридные: комбинация слов и символов, как раз то, что обсуждалось в статье.
